# Sine Wave Line Plot using Matplotlib:

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating sample data
x = np.linspace(0, 10, 50)
y = np.sin(x)  # Calculate the sine values for each x

# Creating a line plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave Line Plot')
plt.grid(True)  # Add a grid to the plot
plt.show()
```

# Scatter Plot using Matplotlib:

```python
import matplotlib.pyplot as plt

# Sample data
x = [1.5, 2, 3.4, 4.3, 5]
y = [11, 15, 14, 17, 19.8]

# Creating a scatter plot
plt.scatter(x, y, marker='o', color='purple', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.legend()
plt.grid(True)
plt.show()
```

# Bar Plot using Seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
categories = ['Apples', 'Bananas', 'Mangoes', 'Grapes']
values = [30, 45, 22, 60]

# Creating a horizontal bar plot using Seaborn
sns.set(style="whitegrid")  
plt.figure(figsize=(8, 6))  
# Creating the horizontal bar plot
ax = sns.barplot(x=values, y=categories, palette="husl")  
ax.set_xlabel('Count')
ax.set_ylabel('Fruits')
ax.set_title('Fruit Count Bar Plot')

# Display the plot
plt.show()
```

# Pie Chart using Matplotlib:

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 40, 30, 55]
colors = ['lime', 'lightcoral', 'lightskyblue', 'orange']
explode = (0.1, 0, 0, 0)  

# Creating a pie chart
plt.figure(figsize=(5, 5))  
plt.pie(sizes, explode=explode, labels=categories, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')

# Display the plot
plt.show()
```
